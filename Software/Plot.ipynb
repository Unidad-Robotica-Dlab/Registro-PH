```python
import pandas as pd
import matplotlib.pyplot as plt

# Importar el archivo CSV
df = pd.read_csv('/home/manuelm/Documentos/DLAB/Registro-PH/Software/data/prueba1_1s_22-03-2024.csv')

# Mostrar las primeras filas del DataFrame para verificar que se haya importado correctamente
std_dev = df['data'].std()

# Calcular los límites del eje Y usando la desviación estándar
y_min = df['data'].min() - std_dev
y_max = df['data'].max() + std_dev

print(std_dev)

# Plotear los datos usando Pandas con límites en el eje Y basados en la desviación estándar
df.plot(x='tiempo', y='data', kind='line', title=f"prueba1 [1s] [30 mins aprox] | dev.std: "f"{std_dev:1f}", xlabel='tiempo', ylabel='pH', grid=True, ylim=(y_min, y_max))
plt.savefig('/home/manuelm/Documentos/DLAB/Registro-PH/Software/data/prueba1_1s_22-03-2024.png')
# Mostrar el gráfico
plt.show()
```
